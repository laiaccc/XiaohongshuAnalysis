# Performing sentiment analysis on the '笔记标题' column
- Prefer Neutral Sentiments: Titles with neutral sentiments tend to have the highest average 赞藏总量. This suggests that titles that are straightforward, informative, or non-emotive might be more effective in garnering engagement.
- Positive Sentiments are Also Effective: Although not as high as neutral titles, positive titles also achieve a good level of engagement. Titles that convey optimism, excitement, or positivity can be engaging.
- Avoid Negative Sentiments: Titles with negative sentiments have the lowest average 赞藏总量. It's advisable to avoid titles that might convey pessimism, criticism, or negative emotions.

```python
from textblob import TextBlob
import numpy as np
```

```python
# Dropping rows with missing '笔记标题'
data_cleaned = data.dropna(subset=['笔记标题'])

# Function to categorize sentiment
def categorize_sentiment(title):
    analysis = TextBlob(title)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'

# Applying the function to the '笔记标题' column
data_cleaned['Sentiment'] = data_cleaned['笔记标题'].apply(categorize_sentiment)

# Analyzing the relationship between 'Sentiment' and '赞藏总量' (Total Likes and Favorites)
sentiment_group = data_cleaned.groupby('Sentiment')['赞藏总量'].mean().sort_values(ascending=False)

# Preparing the results for the analysis
sentiment_categorization = data_cleaned['Sentiment'].value_counts()
sentiment_group, sentiment_categorization
```

```python
from collections import Counter

# Splitting the hashtags in '话题' column and associating them with '赞藏总量'
# Note: The '话题' column contains multiple hashtags per entry, separated by commas

# Creating a function to split the hashtags and associate them with '赞藏总量'
def split_hashtags(row):
    hashtags = str(row['话题']).split(',')
    return [(hashtag.strip(), row['赞藏总量']) for hashtag in hashtags]

# Applying the function and creating a list of tuples (hashtag, 赞藏总量)
hashtag_list = data_cleaned.dropna(subset=['话题']).apply(split_hashtags, axis=1).sum()

# Creating a DataFrame from the list
hashtag_df = pd.DataFrame(hashtag_list, columns=['Hashtag', '赞藏总量'])

# Grouping by hashtag and calculating the average '赞藏总量' for each hashtag
hashtag_engagement = hashtag_df.groupby('Hashtag')['赞藏总量'].mean().sort_values(ascending=False)

# Identifying the top hashtags in terms of average '赞藏总量'
top_hashtags = hashtag_engagement.head(10)

top_hashtags
```

```python
#plotting the most popular hashtags

from wordcloud import WordCloud
import matplotlib.pyplot as plt


# Convert the dataframe to a dictionary with hashtags as keys and '赞藏总量' as values
word_freq_dict = hashtag_df['Hashtag'].value_counts().to_dict()

# Initialize the word cloud object
wordcloud = WordCloud(font_path='SourceHanSansSC-VF.otf', width=800, height=400, background_color='white')

# Generate the word cloud using frequencies
wordcloud.generate_from_frequencies(word_freq_dict)

# Display the word cloud using matplotlib
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide the axes
plt.show()
```

```python
# Filtering hashtags relevant to 留学 (study abroad) and analyzing their association with 赞藏总量

# Function to check if '留学' is in the hashtag
def is_relevant_to_study_abroad(hashtag):
    return '留学' in hashtag

# Filtering the DataFrame for hashtags relevant to 留学
study_abroad_hashtags = hashtag_df[hashtag_df['Hashtag'].apply(is_relevant_to_study_abroad)]

# Grouping by hashtag and calculating the average '赞藏总量' for each relevant hashtag
study_abroad_engagement = study_abroad_hashtags.groupby('Hashtag')['赞藏总量'].mean().sort_values(ascending=False)

# Identifying the top hashtag in terms of average '赞藏总量' that includes 留学
top_study_abroad_hashtags = study_abroad_engagement.head(10)

top_study_abroad_hashtags
```

```python
# Analyzing 笔记标题 and 发布时间 within the scope of hashtags relevant to 留学 (study abroad) for insights on increasing followers

# Filtering the original dataset for posts with 留学-related hashtags
study_abroad_data = data_cleaned[data_cleaned['话题'].str.contains('留学', na=False)]

# Calculating the difference between '当前粉丝数' and '发文时粉丝数' to get the increase/decrease in followers
study_abroad_data['Followers Change'] = study_abroad_data['当前粉丝数'] - study_abroad_data['发文时粉丝数']

# Grouping data by '笔记标题' and calculating the average followers change
title_followers_change = study_abroad_data.groupby('笔记标题')['Followers Change'].mean().sort_values(ascending=False)

# Grouping data by '发布时间' (hour of the day) and calculating the average followers change
study_abroad_data['Hour of Day'] = study_abroad_data['发布时间'].dt.hour
time_followers_change = study_abroad_data.groupby('Hour of Day')['Followers Change'].mean().sort_values(ascending=False)

# Preparing the top results for both analyses
top_titles_for_followers = title_followers_change.head(10)
top_times_for_followers = time_followers_change.head(10)

top_titles_for_followers, top_times_for_followers
```
